#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in Exercise 14) (★★★)

Print both values for each variable.

```python
# TODO
# Iterar sobre cada población del "Cinturón Sur" y calcular la media y varianza
for city in south_belt_populations:
    # Filtrar los datos por ciudad
    city_data = south_belt_subset[south_belt_subset['level5'] == city]
    
    # Calcular las medias
    means = city_data[['price', 'rooms', 'surface', 'bathrooms']].mean()
    
    # Calcular las varianzas
    variances = city_data[['price', 'rooms', 'surface', 'bathrooms']].var()
    
    # Imprimir los resultados para la ciudad actual
    print(f"Resultados para {city}:")
    for var in ['price', 'rooms', 'surface', 'bathrooms']:
        print(f"  Variable: {var}")
        print(f"    Media: {means[var]}")
        print(f"    Varianza: {variances[var]}")
    print("\n")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO

# Función para normalizar precios usando Min-Max Normalization
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

# Crear un diccionario para almacenar los precios normalizados de cada población
normalized_prices = {}

# Normalizar los precios para cada ciudad
for city in south_belt_populations:
    city_data = south_belt_subset[south_belt_subset['level5'] == city]
    normalized_prices[city] = min_max_normalize(city_data['price'])

# Configurar el gráfico con 4 histogramas en la misma figura
plt.figure(figsize=(10, 6))

# Graficar los histogramas para cada ciudad
for city in south_belt_populations:
    plt.hist(normalized_prices[city], bins=20, alpha=0.5, label=city)

# Añadir etiquetas y título
plt.title('Histograma Normalizado de Precios por Población (Cinturón Sur de Madrid)')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el gráfico
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO


# Definir las poblaciones del Cinturón Sur de Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original para obtener el subconjunto con las poblaciones del Cinturón Sur
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)].copy()  # Usamos .copy() para evitar la advertencia

# Crear una nueva columna "pps" (precio por metro cuadrado) usando .loc para evitar la advertencia
south_belt_subset.loc[:, 'pps'] = south_belt_subset['price'] / south_belt_subset['surface']

# Filtrar los datos por las ciudades "Getafe" y "Alcorcón"
getafe_data = south_belt_subset[south_belt_subset['level5'] == 'Getafe']
alcorcon_data = south_belt_subset[south_belt_subset['level5'] == 'Alcorcón']

# Calcular la media del precio por metro cuadrado para cada ciudad
mean_pps_getafe = getafe_data['pps'].mean()
mean_pps_alcorcon = alcorcon_data['pps'].mean()

# Imprimir los resultados
print(f"Precio medio por metro cuadrado en Getafe: {mean_pps_getafe} EUR/m²")
print(f"Precio medio por metro cuadrado en Alcorcón: {mean_pps_alcorcon} EUR/m²")

##Parece que el precio medio por metro cuadrado en Getafe es menor al de Alcorcón. Lo que indica una relación precio/superficie más barata en Getafe, que quiere decir que en Getafe se consiguen casas más grandes en relación a su precio aunque el Alcorcón sean más baratas

```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
# TODO
import matplotlib.pyplot as plt

# Definir las poblaciones del Cinturón Sur de Madrid
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame original para obtener el subconjunto con las poblaciones del Cinturón Sur
south_belt_subset = ds[ds['level5'].isin(populations)].copy()

# Crear una nueva columna "pps" (precio por metro cuadrado)
south_belt_subset.loc[:, 'pps'] = south_belt_subset['price'] / south_belt_subset['surface']

# Crear los subplots para cada población
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Lista para manejar las posiciones en los subplots
positions = [(0, 0), (0, 1), (1, 0), (1, 1)]

# Iterar sobre cada población y crear un scatter plot en el subplot correspondiente
for i, city in enumerate(populations):
    city_data = south_belt_subset[south_belt_subset['level5'] == city]
    ax = axs[positions[i][0], positions[i][1]]
    ax.scatter(city_data['surface'], city_data['pps'], alpha=0.6)
    ax.set_title(f'{city} - Superficie vs Precio/m²')
    ax.set_xlabel('Superficie (m²)')
    ax.set_ylabel('Precio por metro cuadrado (EUR/m²)')

# Ajustar el layout para que no haya solapamientos
plt.tight_layout()
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps

# Map centered on (60 degrees latitude and -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:

from ipyleaflet import Marker, AwesomeIcon
import random

# Definir un diccionario de colores para cada población
colors = {
    'Fuenlabrada': 'blue',
    'Leganés': 'green',
    'Getafe': 'red',
    'Alcorcón': 'purple'
}

# Transformar las columnas de coordenadas (latitud y longitud) en un diccionario por población
locations_by_population = {}
for city in populations:
    city_data = south_belt_subset[south_belt_subset['level5'] == city]
    coordinates = city_data[['latitude', 'longitude']].dropna().values
    locations_by_population[city] = coordinates

# Añadir marcadores al mapa con colores personalizados usando AwesomeIcon
for city, coordinates in locations_by_population.items():
    for coord in coordinates:
        lat, lon = float(coord[0].replace(',', '.')), float(coord[1].replace(',', '.'))
        icon = AwesomeIcon(name="home", marker_color=colors[city], icon_color="white")
        marker = Marker(location=(lat, lon), draggable=False, icon=icon)
        map.add_layer(marker)

map
```
